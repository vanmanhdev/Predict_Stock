```python
import vnquant.data as dt
import pandas as pd

symbols = ['STB', 'SSI', 'TCB', 'VPB', 'HPG', 'MBB', 'POW', 'CTG', 'TCH', 'HDB', 'VHM', 'VRE', 'VIC', 'SBT', 'NVL', 'PDR', 'BID', 'VNM', 'PLX', 'TPB', 'MWG', 'FPT', 'PNJ', 'MSN', 'KDH', 'GAS', 'BVH', 'VJC', 'VCB', 'REE']
all_data = []

for symbol in symbols:
    loader = dt.DataLoader(symbols=[symbol],
                           start="2023-01-01",
                           end="2024-08-20",
                           minimal=False,
                           data_source="cafe")
    data = loader.download()
    data.reset_index(inplace=True)
    data['symbol'] = symbol
    data = data[['date', 'open', 'high', 'low', 'close', 'volume_match', 'symbol']]
    data.rename(columns={'volume_match': 'volume'}, inplace=True)
    all_data.append(data)

combined_data = pd.concat(all_data, ignore_index=True)

combined_data.to_csv('/content/gdrive/MyDrive/SIC_share/price.csv', index=False)
```
